```python
#根據提問回答 該產業去年累計營收和平均營收的關係

import numpy as np

industry = input('請輸入產業類別(產業別: 光電業、 其他、 其他電子業、 化學工業、半導體業、 塑膠工業、 存託憑證、 居家生活、 建材營造、 數位雲端、 橡膠工業、 水泥工業、汽車工業、 油電燃氣業、 玻璃陶瓷、 生技醫療業、 紡織纖維、 綠能環保、航運業、觀光餐旅、 貿易百貨、 資訊服務業、 通信網路業、 造紙工業、運動休閒、金融保險業、 鋼鐵工業、電器電纜、 電子通路業、 電子零組件業、電機機械、電腦及週邊設備業、 食品工業):')
filtered_df = df[df['產業別'] == industry]

row_count = filtered_df.shape[0]
df_sorted = filtered_df.sort_values(by='累計營業收入-去年累計營收', ascending=False)
number=input(f'請輸入好奇名次(總共有 {row_count} 名)')
company_name = df_sorted.iloc[int(number)-1]['公司名稱']
company_money = df_sorted.iloc[int(number)-1]['累計營業收入-去年累計營收']
total = 0

for i in range(row_count):
    revenue = filtered_df.iloc[i]['累計營業收入-去年累計營收']
    total = total+revenue
    
average = total/row_count

if company_money>average:
    money=company_money-average
    print(industry,'中',company_name,'的去年累計營收為',company_money,'元')
    print(industry,f"去年平均營收為: {average}",'元，',company_name,'高於平均',money,'元')
else:
    money=average-company_money
    print(industry,'中',company_name,'的去年累計營收為',company_money,'元')
    print(industry,f"去年平均營收為: {average}",'元',company_name,'低於平均',money,'元')
```
